```python
import sys
import io

company_description=''
i=0

# ticker_list=['AA']

for ticker in ticker_list:
    try:
        print('Ticker:',ticker)
        url = 'https://finance.yahoo.com/quote/'+ticker+'/profile?p='+ticker

        r=requests.get(url)

        soup=BeautifulSoup(r.content,"html.parser")

        inc_data = soup.find_all("div",{"id":"Col1-0-Profile-Proxy"})    


        company_description = str(inc_data[0].text.encode('utf-8'))
        company_description = company_description.replace("\x92","")
        company_description = company_description.replace("company",ticker)
        company_description = company_description.replace("Company",ticker)

        company_description = company_description 
        print(company_description)

        url_news = 'https://finance.yahoo.com/quote/'+ticker+'?p='+ticker

        r_news=requests.get(url_news)
        soup_news=BeautifulSoup(r_news.content,"html.parser")

        inc_data_news = soup_news.find_all("div",{"data-test-locator":"mega"})    

        news_item = ''
        print('News\n')
        for item in inc_data_news:
            news_item = news_item+ str(item.text.encode('utf-8'))

        print(news_item)

        with open('company_'+ticker, "w") as fh:    
            fh.write('Ticker:'+ticker+'\n')
            fh.write(str(company_description))
            fh.write(str(news_item.encode('utf-8')))
            fh.write('\n'+'*'*30 )

        fh.close()
        i = i+1
    except:
        continue
```
