```python
for ticker in ticker_list:
    try:
        csv_file = os.path.join(save_path, 'Stock_'+ticker+'.csv')
        df = pd.read_csv(csv_file,index_col='Date',parse_dates=True)
        df['Price Return']=""
        df['Price Return'].fillna(0,inplace=True)
        df['Price Return']=df['Adj Close'] /df['Adj Close'].shift(1)-1
        df['Price Return Level']=(df['Price Return']+1).cumprod()

        print('==========================\n')
        print('Getting price data for:')
        print(ticker)

        print(df.head())
        df.to_csv('Stock_'+ticker+'.csv')
        
    except:
        continue
```

```python
# Correlation with DDD
save_path = 'C:/Users/Daniel/Documents/Yewno/'

df_corr = pd.DataFrame(columns=['Ticker', 'Price Return Corr', 'Price Return Level Corr'])

for ticker in ticker_list:
    df_DDD = pd.read_csv(os.path.join(save_path, 'Stock_DDD.csv'),index_col='Date',parse_dates=True)
    
    try:
        csv_file = os.path.join(save_path, 'Stock_'+ticker+'.csv')
        df = pd.read_csv(csv_file,index_col='Date',parse_dates=True)
        df['Price Return']=""
        df['Price Return'].fillna(0,inplace=True)
        df['Price Return']=df['Adj Close'] /df['Adj Close'].shift(1)-1
        df['Price Return Level']=(df['Price Return']+1).cumprod()
        df_corr['Ticker']=ticker
        
        print('Correlation in Price Return with '+ticker)
        print(df_DDD['Price Return'].corr(df['Price Return']))
        df_corr['Price Return Corr']=df_DDD['Price Return'].corr(df['Price Return'])

        print('Correlation in Price Return Level with '+ticker)
        print(df_DDD['Price Return Level'].corr(df['Price Return Level']))
        df_corr['Price Return Level Corr']=df_DDD['Price Return Level'].corr(df['Price Return Level'])

        
        print('==========================\n')
        
    except:
        continue

# df_corr['Price Return Level']        
# df_corr.to_csv('Correlation_with_DDD.csv')
```
